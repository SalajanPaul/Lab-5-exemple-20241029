```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_path = "StudentPerformanceFactors.csv"
data = pd.read_csv(data_path)

plt.figure(figsize=(10, 6))
scatter = sns.regplot(x='Hours_Studied', y='Exam_Score', data=data, 
                      scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})

plt.title('Exam score vs Hours_Studied')
plt.xlabel('Hours_Studied')
plt.ylabel('ExamScore')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

data_path = "StudentPerformanceFactors.csv"
data = pd.read_csv(data_path)

X = data[['Hours_Studied', 'Attendance']].values
y = data['Exam_Score'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

nn_model = MLPRegressor(
    hidden_layer_sizes=(100,),
    activation='relu',          
    solver='adam',              
    max_iter=500,               
    random_state=42             
)

nn_model.fit(X_train, y_train)

y_pred_nn = nn_model.predict(X_test)

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_nn, alpha=0.5, color='red', label='Predicted')
plt.scatter(y_test, y_test, alpha=0.5, color='blue', label='Actual')

plt.plot(y_test, y_test, color='green', linewidth=2)

plt.title('Neural Network Predicted Exam Scores vs. Actual Exam Scores')
plt.xlabel('Actual Exam Score')
plt.ylabel('Predicted Exam Score')
plt.legend()
plt.show()
```

```python
r2_score = nn_model.score(X_test, y_test)
print(f"R Score: {r2_score}")
```
